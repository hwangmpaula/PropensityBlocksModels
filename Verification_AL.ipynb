```python
al = gpd.read_file("tl_2019_01_tract/tl_2019_01_tract.shp")
al
```

```python
al['GEOID']
```

```python
def remove0_(geoid):
    if geoid.startswith('0'):
        geoid= geoid.replace('0','', 1)
    else:
        pass
    return geoid
```

```python
al['copy_index'] = al.index
al
```

```python
al_list = []

for index, i in enumerate(zip(al['copy_index'], al['GEOID'])):
    copy_index= i[0]
    GEOID = remove0_(i[1])
    
    al_list.append({'copy_index' : copy_index,
                       'GEOID': GEOID})
```

```python
al_list
```

```python
al_list = pd.DataFrame(al_list, columns = ['copy_index', 'GEOID'])
al_list
```

```python
pd.set_option('display.max_columns', None)
```

```python
al_list = pd.merge(al, al_list, on='copy_index')
al_list
```
