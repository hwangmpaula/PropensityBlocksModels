```python
al = gpd.read_file("shapefiles_blocks/tl_2019_01_tabblock10/tl_2019_01_tabblock10.shp")
al
```

```python
al['copy_index'] = al.index
al.head(2)
```

```python
def remove0_(geoid):
    if geoid.startswith('0'):
        geoid= geoid.replace('0','', 1)
    else:
        pass
    return geoid
```

```python
geoid_list = []

for index, i in enumerate(zip(al['copy_index'], al['GEOID10'])):
    copy_index= i[0]
    GEOID = remove0_(i[1])
    
    geoid_list.append({'copy_index' : copy_index,
                       'GEOID': GEOID})
```

```python
geoid_list = pd.DataFrame(geoid_list, columns = ['copy_index', 'GEOID'])
geoid_list.head(3)
```

```python
new_al = pd.merge(al, geoid_list, left_on='copy_index', right_on= 'copy_index', how = 'left')
new_al
```

```python
new_al = new_al[['GEOID','geometry']]
new_al.head()
```

```python
new_df
```

```python
new_df['copy_index'] = new_df.index
new_df.head(2)
```

```python
def control0_(geoid):
    geoid = geoid.replace('#_', '')
    geoid = geoid.replace(' ', '')
    geoid = geoid
    if geoid.startswith('0'):
        geoid= geoid.replace('0','', 1)
    else:
        pass
    return geoid
```

```python
geoid2_list = []

for index, i in enumerate(zip(new_df['copy_index'], new_df['GEOID'])):
    copy_index= i[0]
    GEOID = control0_(i[1])
    
    geoid2_list.append({'copy_index' : copy_index,
                       'GEOID': GEOID})
```

```python
geoid2_list = pd.DataFrame(geoid2_list, columns = ['copy_index', 'GEOID'])
geoid2_list.head(3)
```

```python
new_df = pd.merge(new_df, geoid2_list, left_on='copy_index', right_on= 'copy_index', how = 'left')
new_df
```

```python
new_df = new_df.rename(columns={"GEOID_y": "GEOID"})
new_df.head()
```
